```python
import pandas as pd
import matplotlib.pyplot as plt
survey_data = pd.read_csv("first_data_test.csv")
survey_data_drop = survey_data.drop(1) # drops row 2
survey_data_drop = survey_data_drop.drop(0)
```

```python
display(survey_data_drop)
```

**Descriptive Statistics for Generative AI Survey Respondents**

1. Context 1, Brainstorm Question 1

```python
# make a bar chart/histogram/graph for amount of people who selected which response
response_count_c1b1 = survey_data_drop['C1_brainstorm_1'].value_counts()
data = """
definitely_allows
probably_allows
might_allow
might_prohibt
probably_prohibt
definitely_prohibit
"""
q1 = data.split()
# plot the graph
response_count_c1b1.plot(kind='hist')
plt.xlabel("Responses")
plt.ylabel("Frequency")
plt.title("C1_brainstorm_1 response frequency")

f, ax = plt.subplots()
ax.set_xticklabels(q1)

plt.show()
```
