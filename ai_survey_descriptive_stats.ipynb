```python
from scipy.stats import norm
def bayesian_test(data, prior_mu, prior_sigma):
    n = len(data)                           # number of observations in data
    data_mean = np.mean(data)               # sample mean calculation

    prior_precision = 1 / prior_sigma**2    # convert into precision??
    data_precision = 1 / np.var(data)

    posterior_precision = prior_precision + n * data_precision      # calculate precision by summing prior precision and precision of data x number observations; represents how much information the posterior has
    posterior_variance = 1 / posterior_precision                    # converts posterior precision back to variance
    posterior_mu = (prior_mu * prior_precision + data_mean * n * data_precision) / posterior_precision          # mean of posterior distribution
    posterior_sigma = np.sqrt(posterior_variance)                   # std of posterior distribution from variance

    return posterior_mu, posterior_sigma     # return mean and std of posterior distribution

data = reshaped_df1_filtered['value'].values
prior_mu = 3                                 # estimation of average response (prior)
prior_sigma = 2                              # level of uncertainty in prior

posterior_mu, posterior_sigma = bayesian_test(data, prior_mu, prior_sigma)

print(f"Posterior mean: {posterior_mu}, Posterior sigma: {posterior_sigma}")
# posterior mean represents average response estimated by model
# posterior sigma is estimated std of responses based on posterior (how much variation)



```

```python
def bayesian_height_test(low, high, height_data):
    mean_h = np.mean(height_data)
    variance_h = 1
    std_h = np.sqrt(variance_h)

    probability_h_low = norm.cdf(low, mean_h, std_h)
    probability_h_high = norm.cdf(high, mean_h, std_h)
    probability_h = probability_h_high - probability_h_low

    likelihood = norm.pdf(h, mean_h, std_h)

    denominator = 0
    step_size = 0.1
    for i in np.arange(low, high+step_size, step_size):
        ph_step_low = norm.cdf(i, mean_h, std_h)
        ph_step_high = norm.cdf(i+step_size, mean_h, std_h)
        ph_step = ph_step_high - ph_step_low

        likelihood_step = norm.pdf(i, mean_h, std_h)
        denominator += ph_step * likelihood_step

    return (probability_h * likelihood) / denominator


# height_data = [10,20,30,50,80,90,170,30,25]
# result = bayesian_height_test(30,50,height_data,150)
# print(result)

height_data = np.random.normal(170, 10, 1000)
result = bayesian_height_test(160,180,height_data,175)
print(result)
```
